```python
def plot_tracks(cruise_id, cols, labels=None):
    fig, axes = plt.subplots(len(cols), 1, figsize=(20, 7*len(cols)), sharex=True)
    if labels is None:
        labels = cols
    for (i, col) in enumerate(cols):
        plot_track_data(axes[i], col, cruise_id, label=labels[i])

def plot_track_data(ax, col, cruise_id, label=None):
    cruise = df.loc[cruise_id].sort_values(by='time_avg')
    sal_cruise = cruise.pipe(viz.clean_indicies_for_plotting)[col]
    if label is None:
        label=col
    ax.set_title(label, fontsize=20)
    sal_cruise.plot(ax=ax)
    plt.xticks(rotation=45)

#from chatgpt
def analyze_time_intervals(df, time_column):
    """
    Analyze the regularity of time sampling in a pandas dataframe.

    Parameters:
    - df: pandas DataFrame containing time data.
    - time_column: str, name of the column storing timestamps (as pd.Datetime objects).
    
    Returns:
    - A dictionary with statistical summary of time intervals.
    - Two plots: histogram of time intervals and time series of intervals.
    """
    # Ensure the time column is sorted
    df = df.sort_values(by=time_column)
    
    # Compute time intervals (differences between consecutive timestamps)
    df['time_diff'] = df[time_column].diff().dropna().dt.total_seconds()
    
    # Drop the first row since it will have NaN after diff()
    df = df.dropna(subset=['time_diff'])

    # Compute statistics
    stats = {
        "min_interval (s)": df['time_diff'].min(),
        "max_interval (s)": df['time_diff'].max(),
        "mean_interval (s)": df['time_diff'].mean(),
        "median_interval (s)": df['time_diff'].median(),
        "std_interval (s)": df['time_diff'].std()
    }
    
    # Plot histogram of time intervals
    plt.figure(figsize=(8, 4))
    plt.hist(df['time_diff'], bins=30, edgecolor='black')
    plt.xlabel("Time Interval (seconds)")
    plt.ylabel("Frequency")
    plt.title("Histogram of Time Intervals")
    plt.show()
    
    # Plot time series of intervals
    plt.figure(figsize=(10, 4))
    plt.plot(df[time_column], df['time_diff'], marker='o', linestyle='-')
    plt.xlabel("Timestamp")
    plt.ylabel("Time Interval (seconds)")
    plt.title("Time Intervals Over Time")
    plt.xticks(rotation=45)
    plt.show()

    return stats
```

```python
stats = analyze_time_intervals(df, "time_avg")
print(stats)
```

## Check distances between ship coordinates and nearest satellite samples

```python
# check distances 
lats = df.index.get_level_values('lat_005').to_frame()
lons = df.index.get_level_values('lon_005').to_frame()
print(lats)
```

```python
# Using GeoPandas with Shapely (built on top of GEOS C library)
import geopandas as gpd
import numpy as np

def calculate_distances_geopandas(df, dsnames):
    # Create geometry for first set of points
    geometry1 = gpd.points_from_xy(lons['lon_005'], lats['lat_005'])
    gdf1 = gpd.GeoDataFrame(geometry=geometry1, crs="EPSG:4326")

    dists = []
    for dsname in dsnames:
        # Create geometry for second set of points
        geometry2 = gpd.points_from_xy(df['lon_' + dsname], df['lat_' + dsname])
        gdf2 = gpd.GeoDataFrame(geometry=geometry2, crs="EPSG:4326")
        
        # Transform to a projected CRS for more accurate distance calculations
        gdf1 = gdf1.to_crs("EPSG:3857")  # Web Mercator projection
        gdf2 = gdf2.to_crs("EPSG:3857")
        
        # Calculate distances in kms
        dists.append(np.array([pt1.distance(pt2) / 1000 for pt1, pt2 in zip(gdf1.geometry, gdf2.geometry)]))
    
    
    return pd.DataFrame(np.array(dists).T, index=df.index, columns=dsnames)

distances = calculate_distances_geopandas(df, ['chlgc', 'soda', 'sshcm', 'sstcci', 'ssscci'])
```
